```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.isnull().sum()
```

## DROP Unnamed column

```python
df = df.drop("Unnamed: 0" , axis = 1)
print(df.head())
```

# gender distribution

```python
plt.figure(figsize= (3,4))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from the above chart we have analysed that:
# the number of females in the data is more than the number og males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(gb, annot=True, fmt=".1f", cmap='coolwarm')
plt.xticks(rotation=45, ha='right')
plt.title("Relationship between Parent's Education and Student's Score", pad=20)
plt.tight_layout()
plt.show()
```

```python
#from the above chart we have concluded:
# Parents education have an impact on their child's education.
```

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(gb, annot=True, fmt=".2f", cmap='coolwarm', cbar_kws={'label': 'Score Mean'})
plt.xticks(rotation=45, ha='right')
plt.title("Relationship between Parent's Education Level and Student's Scores", pad=20)
plt.xlabel("Subject Scores")
plt.ylabel("Parent's Education Level")
plt.tight_layout()
plt.show()
```

```python
# from the above chart we have analyzed that:
# Marital status of the parents, have no impact or negligible impact.
```
